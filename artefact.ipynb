# 3. Discretizing Alcohol Content

```python
# Categorization function
def discretise_alcohol(content, avg, stddev):
    if content < avg - stddev:
        return 'low'
    elif content > avg + stddev:
        return 'high'
    else:
        return 'mid'
```

```python
# Mean and standard deviation
red_avg = red_wine['alcohol'].mean()
red_std = red_wine['alcohol'].std()
# Applying the function
red_wine['alcohol_cat'] = red_wine['alcohol'].apply(discretise_alcohol, args=(red_avg, red_std))
```

```python
white_avg = white_wine['alcohol'].mean()
white_std = white_wine['alcohol'].std()
white_wine['alcohol_cat'] = white_wine['alcohol'].apply(discretise_alcohol, args=(white_avg, white_std))
```

# 6. "isSweet" Binary Variable

```python
# Define a threshold for sweet wine
mid_ressug_red = red_wine['residual sugar'].median()
mid_ressug_white = white_wine['residual sugar'].median()
red_wine['isSweet'] = (red_wine['residual sugar'] > mid_ressug_red).astype(int)
white_wine['isSweet'] = (white_wine['residual sugar'] > mid_ressug_white).astype(int)
```

# 8. Machine Learning Model Experimentation

```python
# Binary quality variable for wine
red_wine['quality_binary'] = (red_wine['quality'] >= 6).astype(int)
white_wine['quality_binary'] = (white_wine['quality'] >= 6).astype(int)
```

```python
from sklearn.model_selection import train_test_split

X_red = red_wine.drop(['quality', 'quality_binary'], axis=1)
y_red = red_wine['quality_binary']
X_train_red, X_test_red, y_train_red, y_test_red = train_test_split(X_red, y_red, test_size=0.2, random_state=42)

X_white = white_wine.drop(['quality', 'quality_binary'], axis=1)
y_white = white_wine['quality_binary']
X_train_white, X_test_white, y_train_white, y_test_white = train_test_split(X_white, y_white, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
```

```python
# Drop the 'alcohol_cat' column from the feature set since it's a string
X_train_red = X_train_red.drop('alcohol_cat', axis=1)
X_test_red = X_test_red.drop('alcohol_cat', axis=1)

X_train_white = X_train_white.drop('alcohol_cat', axis=1)
X_test_white = X_test_white.drop('alcohol_cat', axis=1)
```

```python
rfc_red = RandomForestClassifier(random_state=42)
rfc_red.fit(X_train_red, y_train_red)
y_pred_red = rfc_red.predict(X_test_red)

print(classification_report(y_test_red, y_pred_red))
print("Accuracy:", accuracy_score(y_test_red, y_pred_red))
```

```python
rfc_white = RandomForestClassifier(random_state=42)
rfc_white.fit(X_train_white, y_train_white)
y_pred_white = rfc_white.predict(X_test_white)

print(classification_report(y_test_white, y_pred_white))
print("Accuracy:", accuracy_score(y_test_white, y_pred_white))
```

# 9. K-fold Cross-validation

```python
from sklearn.model_selection import cross_val_score
```

```python
cv_scores_red = cross_val_score(rfc_red, X_train_red, y_train_red, cv=5)
print("CV Scores: ", cv_scores_red)
print("Mean CV Score: ", cv_scores_red.mean())
print("Standard Deviation in CV Scores: ", cv_scores_red.std())
```

```python
cv_scores_white = cross_val_score(rfc_white, X_train_white, y_train_white, cv=5)
print("CV Scores: ", cv_scores_white)
print("Mean CV Score: ", cv_scores_white.mean())
print("Standard Deviation in CV Scores: ", cv_scores_white.std())
```

# 10. Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20],
}

grid_search = GridSearchCV(rfc_red, param_grid, cv=5, n_jobs=-1)

grid_search.fit(X_train_red, y_train_red)

print("Best Parameters: ", grid_search.best_params_)
print("Best CV Score: ", grid_search.best_score_)
```

```python
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20],
}

grid_search = GridSearchCV(rfc_white, param_grid, cv=5, n_jobs=-1)

grid_search.fit(X_train_white, y_train_white)

print("Best Parameters: ", grid_search.best_params_)
print("Best CV Score: ", grid_search.best_score_)
```

# 11. Model Evaluation with Test Set

```python
best_model = grid_search.best_estimator_
y_pred_red = best_model.predict(X_test_red)
print(classification_report(y_test_red, y_pred_red))
print("Test Set Accuracy: ", accuracy_score(y_test_red, y_pred_red))
```

```python
best_model = grid_search.best_estimator_
y_pred_white = best_model.predict(X_test_white)
print(classification_report(y_test_white, y_pred_white))
print("Test Set Accuracy: ", accuracy_score(y_test_white, y_pred_white))
```
